# GTO vs Grid

Grid法の計算結果とGTOによる計算結果を比較している。実部のみプロットした。波動関数の振幅としては、PFのみが非常に大きくそれ以外は非常に小さい。s on Hを加えたことによりL=5,7の相互作用領域におけるdriven term由来の振る舞いを表現している。しかし、最初の振幅を評価するだけで、第二の波は全く評価できていない。適当な複素基底関数を導入することでこの問題が解決するかもしれない。

gto_null : p,f-GTO on center

gto_3    : GTO_null + s_on_H(orbital exponent = 3)

gto_39   : GTO_null + s_on_H(orbital exponent = 3,9)

```python
base = "../../../../cbasis/calc/h2plus/add_s_on_H_l/out/"
df_psi_gto = pd.read_csv(base + "null.psi.csv", index_col=None)
df_psi_gto_3 = pd.read_csv(base + "3.psi.csv", index_col=None)
df_psi_gto_39 = pd.read_csv(base + "3_9.psi.csv", index_col=None)
fig, axes = plt.subplots(4,1)
fig.set_size_inches(5, 10)
for (L,ax) in zip([1,3,5,7], axes):
    ax.set_title("L={0}".format(L))
    ax.set_xlim(0, 20)
    ax.plot(df_psi_gto["r"], -df_psi_gto["re_{0}".format(L)],      label="GTO_null")
    ax.plot(df_psi_gto_3["r"], -df_psi_gto_3["re_{0}".format(L)],  label="GTO_3")
    ax.plot(df_psi_gto_39["r"], -df_psi_gto_39["re_{0}".format(L)],label="GTO_39")
    ax.plot(df_psi1["r"],    df_psi1["re_{0}".format(L)], "k--", label="grid")        
    ax.legend()
```
