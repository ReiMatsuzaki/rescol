```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

# plot real and imaginary

```python
df_psi1 = pd.read_csv("psi1.dat")
fig, axes = plt.subplots(3,1)
fig.set_size_inches(5, 10)
axes[0].plot(df_psi1["r"], df_psi1["re_1"])
axes[0].plot(df_psi1["r"], df_psi1["im_1"])
axes[1].plot(df_psi1["r"], df_psi1["re_3"])
axes[1].plot(df_psi1["r"], df_psi1["im_3"])
axes[2].plot(df_psi1["r"], df_psi1["re_5"])
axes[2].plot(df_psi1["r"], df_psi1["im_5"])
```

# plot only real

```python
df_psi1 = pd.read_csv("psi1.dat")
fig, axes = plt.subplots(1,1)
axes.set_xlim(0, 40)
axes.set_ylim(-0.1, 0.1)
axes.plot(df_psi1["r"], df_psi1["re_1"], label="L=1")
axes.plot(df_psi1["r"], df_psi1["re_3"], label="L=3")
axes.plot(df_psi1["r"], df_psi1["re_5"], label="L=5")
axes.plot(df_psi1["r"], df_psi1["re_7"], label="L=7")
plt.legend()
```

# GTO vs Grid

GridGTOPFs on HL=5,7driven term

gto_null : p,f-GTO on center

gto_3    : GTO_null + s_on_H(orbital exponent = 3)

gto_39   : GTO_null + s_on_H(orbital exponent = 3,9)

```python
base = "../../../../cbasis/calc/h2plus/add_s_on_H_l/out/"
df_psi_gto = pd.read_csv(base + "null.psi.csv", index_col=None)
df_psi_gto_3 = pd.read_csv(base + "3.psi.csv", index_col=None)
df_psi_gto_39 = pd.read_csv(base + "3_9.psi.csv", index_col=None)
fig, axes = plt.subplots(4,1)
fig.set_size_inches(5, 10)
for (L,ax) in zip([1,3,5,7], axes):
    ax.set_title("L={0}".format(L))
    ax.set_xlim(0, 20)
    ax.plot(df_psi_gto["r"], -df_psi_gto["re_{0}".format(L)],      label="GTO_null")
    ax.plot(df_psi_gto_3["r"], -df_psi_gto_3["re_{0}".format(L)],  label="GTO_3")
    ax.plot(df_psi_gto_39["r"], -df_psi_gto_39["re_{0}".format(L)],label="GTO_39")
    ax.plot(df_psi1["r"],    df_psi1["re_{0}".format(L)], "k--", label="grid")        
    ax.legend()
```
